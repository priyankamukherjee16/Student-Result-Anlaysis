```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/student_scores.csv")
print(df)
```

```python
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#change weekly study hours column

df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("WklyStudyHours","5-10")
df.head()
```

```python
#gender distribution


plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# FROM THE ABOVE CHART WE HAVE ANALYSED THAT
#NO. OF FEMALES IS GREATER THAN NO. OF MALES
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent Education and Student Scores")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE ANALYZED THAT
#EDUCATION OF THE PARENTS HAVE A GOOD IMPACT ON THE STUDENT'S EDUCATION
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent Marital Status and Student's Scores")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE CONCLUDED THAT
#THERE IS NEGLIGIBLE EFFECT ON STUDENT'S SCORES DUE TO THEIR PARENT MARITAL STATUS
```

```python
gb2=df.groupby("NrSiblings").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb2,annot=True)
plt.title("Realtionship between No.Of Siblings and Student's Scores")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE CONCLUDED THAT
#THERE IS NEGLIGIBLE EFFECT ON STUDENT'S SCORES DUE TO NO.OF SIBLINGS
```

```python
gb3=df.groupby("PracticeSport").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb3)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb3,annot=True)
plt.title("Relationship between Practice Sports and Students Marks")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE CONCLUDED THAT
#THERE IS NO EFFECT ON STUDENT'S SCORES DUE TO PRACTICE SPORTS
```

```python
gb4=df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb4)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb4,annot=True)
plt.title("Relationship between WklyStudyHours and Student's Scores ")
plt.show()
```

```python
#FROM THE ABOVE CHART WE HAVE CONCLUDED THAT
#GREATER THAN 10 WklyStudyHours HAVE A GOOD IMPACT ON STUDENT'S SCORES
```
